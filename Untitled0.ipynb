```python
import json
from google.colab import drive
drive.mount('/drive/', force_remount=True)
ruta = '/drive/PrimeraEntrega'
```

```python

def registro(BD):
 
    usuario=input("Ingrese el nombre de usuario:")
    if usuario in BD.keys():
      print("Ese usuario ya existe")
    else:
      password=input("Ingrese la contraseña:")
      BD[usuario] = password 
```

```python
def leerData(BD):
  print("La info almacenada en la base de datos es:")
  for usuario, password in BD.items():
    print("usuario:", usuario, "password:", BD[usuario])
   
```

```python
def guardarArchivo(BD):
  with open(ruta + "/archivo.txt", "w") as file:
    json.dump(BD, file, indent=4)
```

```python
def login(BD):
    usuario=input("Ingrese su usuario:")
    if usuario in BD.keys():
      password=input("Ingrese su contraseña:")
      if(BD[usuario] == password):
        print("Has ingresado  correctamente")
      else:
        print("Contraseña incorrecta")
    else:
      print("Usuario inexistente")
```
